```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
file_path = "koopman_training_data.npy"
data = np.load(file_path)
data.shape
```

```python
control = data[:,:,:7]
state = data[:,:,7:]
```

```python
flatten_state = state.transpose(1,0,2).reshape(-1, 18)
flatten_control = control.transpose(1,0,2).reshape(-1, 7)
print(flatten_state.shape)
print(flatten_control.shape)
```

```python
for i in range(7):
    plt.plot(flatten_control[:50, i], 'o', label=f'Control {i+1}')
plt.xlabel('Control Input')
plt.ylabel('Control Input')
plt.title('Control Inputs')
plt.legend()
plt.grid()
plt.show()

for i in range(18):
    plt.plot(flatten_state[:50, i], 'o', label=f'State {i+1}')
plt.xlabel('State Variable')
plt.ylabel('State Variable')
plt.title('State Variables')
plt.legend()
plt.grid()
plt.show()
```
